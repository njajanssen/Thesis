```python
from mmnl import MMNL
import numpy as np
import matplotlib.pyplot as plt
import pickle
import pandas as pd
```

```python
def load_data(path):
    dat = np.load(path)
    X = dat[:, :-1]
    Y = np.reshape(dat[:, -1], (-1, 1))
    return X, Y
X, Y = load_data('data/data.npy')
```

Load data

```python
%%time
infile = open('./data/300_dgp.p', 'rb')
big_dict = pickle.load(infile)
Y_dgp = big_dict['theta: [ 1.5  1.  -1.1  0.8  0.1  1.2]']
print(Y_dgp.shape)
```

```python
qmc = MMNL(X,Y_dgp[:,0],75,3,method='QMC')
solution = qmc.solver()
```

```python
qmc.log_likelihood(np.array([ -1.95246613,   0.78251688, -21.00889558,  12.27569976,
        11.90995196,   0.02778854]))
```

```python
%%time
print(Y_dgp[:,0])
result = []
for i in range(100):
    print('iteration: %i' %(i+1))
    qmc = MMNL(X,Y_dgp[:,i],75,3,method='QMC')
    solution = qmc.solver()
    result.append(solution.x)
pickle.dump( result, open( str(100)+"_results.p", "wb" ) )
```
