```python
from mmnl import MMNL
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize
```

```python
def load_data(path):
    dat = np.load(path)
    X = dat[:, :-1]
    Y = np.reshape(dat[:, -1], (-1, 1))
    return X, Y
X, Y = load_data('data/data.npy')
```

```python
bm = MMNL(X,Y,8,3,'BQMC')
qmc = MMNL(X,Y,100,3,method='QMC')
theta0 = np.array([0.,0.,0.,1.,1.,1.])
```

```python

min = np.inf
a = 7

llq = qmc.log_likelihood(theta0)
print(llq)
for a in range(1,10):
    for b in range(1,10):
        for c in range(1,10):

            bm.set_w(np.array([1.,a,b,c]))
            llb = bm.log_likelihood(theta0)
            if llb < min:
                min = llb
                print('loglig: %f, w: %f,%f,%f, new min log' % (llb,a,b,c))
            else:
                print('loglig: %f, w: %f,%f,%f' % (llb, a, b, c))
```
